```python
import pandas as pd
from scipy.spatial import distance
import numpy as np
```

```python
feat = pd.read_csv(r'C:\Users\lttso\Desktop\feature_task.csv')
```

```python
conj = pd.read_csv(r'C:\Users\lttso\Desktop\conjunction_task.csv')
```

```python
feat['run_length'] = feat['runLength'] - 1
feat['run_length'] = feat['run_length'].shift(1)
feat = feat[feat['run_length'].notna()]
```

```python
feat
```

```python
conj['run_length'] = conj['runLength'] - 1
conj['run_length'] = conj['run_length'].shift(1)
conj = conj[conj['run_length'].notna()]
```

```python
conj
```

### Delete '[  ]' from each row in coord coluns (for changing to float)

```python
feat['resp_pos'] = feat['resp_pos'].str[1:-1]
feat['closest_same_target'] = feat['closest_same_target'].str[1:-1]
feat['closest_other_target'] = feat['closest_other_taget'].str[1:-1]
feat['closest_blue_dot'] = feat['closest_blue_dot'].str[1:-1]
feat['closest_yellow_dot'] = feat['closest_yellow_dot'].str[1:-1]
```

```python
conj['resp_pos'] = conj['resp_pos'].str[1:-1]
conj['closest_same_target'] = conj['closest_same_target'].str[1:-1]
conj['closest_other_target'] = conj['closest_other_target'].str[1:-1]
conj['closest_dist_same_color'] = conj['closest_dist_same_color'].str[1:-1]
conj['closest_dist_same_shape'] = conj['closest_dist_same_shape'].str[1:-1]
```

### Change type of coord 'str' to 'float'

```python
feat['resp_pos'] = feat['resp_pos'].apply(lambda x: [float(y) for y in x.split(',')])
feat['closest_same_target'] = feat['closest_same_target'].apply(lambda x: [float(y) for y in x.split(',')])
feat['closest_other_target'] = feat['closest_other_target'].apply(lambda x: [float(y) for y in x.split(',')])
feat['closest_blue_dot'] = feat['closest_blue_dot'].apply(lambda x: [float(y) for y in x.split(',')])
feat['closest_yellow_dot'] = feat['closest_yellow_dot'].apply(lambda x: [float(y) for y in x.split(',')])
```

```python
conj['resp_pos'] = conj['resp_pos'].apply(lambda x: [float(y) for y in x.split(',')])
conj['closest_same_target'] = conj['closest_same_target'].apply(lambda x: [float(y) for y in x.split(',')])
conj['closest_other_target'] = conj['closest_other_target'].apply(lambda x: [float(y) for y in x.split(',')])
conj['closest_dist_same_color'] = conj['closest_dist_same_color'].apply(lambda x: [float(y) for y in x.split(',')])
conj['closest_dist_same_shape'] = conj['closest_dist_same_shape'].apply(lambda x: [float(y) for y in x.split(',')])
```

### Make columns with coord as np array

```python
feat['cl_same'] = ((np.array(feat['resp_pos'].values.tolist())
                              - np.array(feat['closest_same_target']
                                         .values
                                         .tolist()))**2).sum(axis=1) ** 0.5
feat['cl_other'] = ((np.array(feat['resp_pos'].values.tolist()) 
                                 - np.array(feat['closest_other_target']
                                            .values
                                            .tolist()))**2).sum(axis=1) ** 0.5
feat['cl_blue'] = ((np.array(feat['resp_pos'].values.tolist()) 
                                 - np.array(feat['closest_blue_dot']
                                            .values
                                            .tolist()))**2).sum(axis=1) ** 0.5
feat['cl_yellow'] = ((np.array(feat['resp_pos'].values.tolist()) 
                                 - np.array(feat['closest_yellow_dot']
                                            .values
                                            .tolist()))**2).sum(axis=1) ** 0.5
```

```python
conj['cl_same'] = ((np.array(conj['resp_pos'].values.tolist())
                              - np.array(conj['closest_same_target']
                                         .values
                                         .tolist()))**2).sum(axis=1) ** 0.5
conj['cl_other'] = ((np.array(conj['resp_pos'].values.tolist()) 
                                 - np.array(conj['closest_other_target']
                                            .values
                                            .tolist()))**2).sum(axis=1) ** 0.5
conj['cl_same_color'] = ((np.array(conj['resp_pos'].values.tolist()) 
                                 - np.array(conj['closest_dist_same_color']
                                            .values
                                            .tolist()))**2).sum(axis=1) ** 0.5
conj['cl_same_shape'] = ((np.array(conj['resp_pos'].values.tolist()) 
                                 - np.array(conj['closest_dist_same_shape']
                                            .values
                                            .tolist()))**2).sum(axis=1) ** 0.5
```

```python
conj
```
